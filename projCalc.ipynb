## Projeto 1 de Cálculo 

```python
df = pd.read_excel('aqmesh-location-data-Sonora Engenharia_1725150.xlsx', parse_dates = True, index_col= 'Project Time (Interval start) (UTC-3)')
display(df)
```

## Limpando o DataFrame

```python
df_1 = df.drop(['Location Name','POD Serial Number',
                'Project Time (Interval end) (UTC-3)',
                'Gas Protocol Version','CO State','NO State','NO2 State',
                'NOx State','O3 State','SO2 State','Particle Protocol Version','PM State',
                'CO PreScaled ugm3','NO PreScaled ugm3','NO2 PreScaled ugm3',
                'NOx PreScaled ugm3','O3 PreScaled ugm3','SO2 PreScaled ugm3',
                'PM Total PreScaled µg/m³','PM10 PreScaled µg/m³','TPC PreScaled count/m³','PM4 PreScaled µg/m³' ],axis = 1)
display(df_1)
```

# Séries Temporais de Cada Gás e particulado em função do tempo



```python
colunas = ['CO Scaled ugm3', 'NO Scaled ugm3', 'NO2 Scaled ugm3', 'NOx Scaled ugm3', 'O3 Scaled ugm3', 'SO2 Scaled ugm3',
           'PM1 PreScaled µg/m³', 'PM1 Scaled µg/m³', 'PM2.5 PreScaled µg/m³',
           'PM2.5 Scaled µg/m³', 'PM4 Scaled µg/m³', 'PM10 Scaled µg/m³',
           'PM Total Scaled µg/m³']
for i in colunas:
    plt.plot(df_1[i],color='red')
    plt.title(i,size= 14)
    plt.show()
    
```

```python
plt.figure(figsize=(16, 8), dpi=150)

df_1['Humidity %RH'].plot(label='Humidade', color = 'blue')
df_1['Temperature C'].plot(label='Temperatura C',color='red')
df_1['Temperature F'].plot(label='Temperatura F',color='orange')

plt.title('Pressure and Humudity')
plt.xlabel('Project Time (Interval start) (UTC-3)')
plt.legend()
```
